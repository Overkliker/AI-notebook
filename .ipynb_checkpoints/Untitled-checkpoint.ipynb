```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': (9, 6)})
eventdf = pd.read_csv(r'C:\Users\kliker\Desktop\event_data_train.csv')
subdf = pd.read_csv(r'C:\Users\kliker\Desktop\submissions_data_train.csv')
pd.set_option('display.max_columns', None)

eventdf['date'] = pd.to_datetime(eventdf.timestamp, unit='s')
eventdf['day'] = eventdf.date.dt.date
subdf['date'] = pd.to_datetime(eventdf.timestamp, unit='s')
subdf['day'] = eventdf.date.dt.date

eventdf.groupby('day').user_id.nunique().plot()

```

```python
eventdf.pivot_table(index='user_id', columns='action', values='step_id', aggfunc='count', fill_value=0).reset_index().hist()
```

```python
gap_data = eventdf[['user_id', 'day', 'timestamp']]\
.drop_duplicates(subset=['user_id', 'day'])\
.groupby('user_id')['timestamp']\
.apply(list).apply(np.diff).values
```

```python
gap_data = pd.Series(np.concatenate(gap_data, axis=0))
```

```python
gap_data / (24 * 60 * 60)
```
